```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
# This makes the plots appear inside the notebook
%matplotlib inline


# Define our derivative
def f(v,t):
    x, y = v;
    return np.array([-x-y,x-y])

# define step size
h = 0.1;
n_steps = 100;

# define our initial conditions.
time = np.zeros(n_steps);
output = np.zeros((n_steps,2));
output[0] = [3,5];

for m in np.arange(1,n_steps,1):
    time[m]=time[m-1]+h;
    output[m]=output[m-1]+h*f(output[m-1],time[m-1]);
    
plt.xlabel("Time")
plt.ylabel("x/y")
plt.plot(time,output)
```

```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from scipy.optimize import fsolve 
# This makes the plots appear inside the notebook
%matplotlib inline

# define step size
h = 0.1;
n_steps = 100;

# define our initial conditions.
time = np.zeros(n_steps);
output = np.zeros((n_steps,2));
output[0] = [3,5];

def func(y,m):
    return y-output[m-1]-h*f(y,0);

for m in np.arange(1,n_steps,1):
    time[m]=time[m-1]+h;
    output[m]=fsolve(func, output[m-1], args = m)
    
plt.xlabel("Time")
plt.ylabel("x/y")
plt.plot(time,output)
```

# Problem 2 Linearization

You're given the nonlinear, first-order, time-invariant ODE $\frac{dy}{dt} = sin(y)$.

1. In the space below, plot the flow diagram for the ODE. 

```python
def dydt(y):
    dydt = math.sin(y)
    return dydt

y_values = np.linspace(-10,10,100)

dydt_values = []
for m in y_values:
    dydt_values.append(dydt(m))
    
# Plot derivative vs y
plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("y")
plt.ylabel("dydt")
plt.plot(y_values,dydt_values);
plt.axhline(y=0, color="black", linestyle="-")
plt.axvline(x=0, color="black", linestyle="-")
```

2. In the space below, type out the equation for the linearized form of our ODE when it is linearized about the lowest, positive-valued, stable steadystate.

**Positive value stable steady point is $\pi$:**

\begin{array}{l}
a = \pi \\
f\left( y \right) = f\left( \pi \right) + \left( {y - \pi} \right)f'\left( \pi \right)\\
f'\left( \pi \right) = \cos \left( \pi \right)\\
f\left( y \right) = \sin \left( y \right) = \sin \left( \pi  \right) + \left( {y - \pi } \right)\cos \left( \pi \right)\\
\sin \left( \pi  \right) = 0\\
\cos \left( \pi  \right) = -1\\
f\left( y \right) =  \left( {-y + \pi } \right)\\
f\left( y \right) =  \left( {\pi - y } \right)\\
\end{array}

3. We want to compare the responses of our linearized model to our original nonlinear model. On a single figure, plot the time response of both models when starting at an initial conditions of $y_0 = 3$. Clearly indicate which lines represent which ODEs.

```python
def dydt1(y,t):
    dydt = math.sin(y)
    return dydt

def dydt2(y,t):
    dydt = (math.pi - y)
    return dydt
from scipy.integrate import odeint

timesteps = np.linspace(0,10,100)

y0 = 3  # the initial condition
y1 = odeint(dydt1, y0, timesteps)
y2 = odeint(dydt2, y0, timesteps)

# Plot the numerical solution
plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("time")
plt.ylabel("$y_1, y_2$")
plt.plot(timesteps, y1,label='Actual');
plt.plot(timesteps, y2,label='Linearized @ $ y = \pi$', color='red', marker = '+', linestyle = 'none',markersize=8);
plt.legend();
```

3. Let's test one more initial condition. On a single figure, plot the time response of both models when starting at an initial conditions of $y_0 = 1$. Clearly indicate which lines represent which ODEs.

```python
timesteps = np.linspace(0,10,100)

y0 = 1  # the initial condition
y1 = odeint(dydt1, y0, timesteps)
y2 = odeint(dydt2, y0, timesteps)

# Plot the numerical solution
plt.rcParams.update({'font.size': 14})  # increase the font size
plt.xlabel("time")
plt.ylabel("$y_1, y_2$")
plt.plot(timesteps, y1,label='Actual');
plt.plot(timesteps, y2,label='Linearized @ $ y = \pi$', color='red', marker = '+', linestyle = 'none',markersize=8);
plt.legend();
```
